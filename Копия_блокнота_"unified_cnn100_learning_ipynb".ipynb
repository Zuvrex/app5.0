```python
def merge_df_with_crop(df_crop, *all_dfs):
    result = []
    for df in all_dfs:
        new_df = pd.merge(df, df_crop, on='id',  how='left')
        result.append(new_df)
    return result
```

```python
dataset2_all_data, dataset2_train, dataset2_val, dataset2_test = merge_df_with_crop(dataset2_crop, dataset2_all_data, dataset2_train, dataset2_val, dataset2_test)
```

```python
crop_columns = ['x1_crop', 'y1_crop', 'x2_crop', 'y2_crop']
dataset1_all_data[crop_columns] = -1
dataset1_train[crop_columns] = -1
dataset1_val[crop_columns] = -1
dataset1_test[crop_columns] = -1
```

```python
def unite_dfs(*all_dfs):
    candidates = []
    id_offset = 0
    for df in all_dfs:
        new_df = df
        if id_offset > 0:
            new_df['id'] = new_df['id'].apply(lambda x: x + id_offset)
        id_offset = new_df['id'].max()+1
        candidates.append(new_df)
    return pd.concat(candidates, ignore_index=True)
```

```python
datasets_united_train = unite_dfs(dataset1_train, dataset2_train)
```

```python
class My_File_Generator(Sequence):
    def __init__(self, dataset_df, target_size=(200,200), batch_size=32, shuffle=True, need_normalize=True):
        self.dataset_df = dataset_df
        self.target_size = target_size
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.need_normalize = need_normalize
        self.on_epoch_end()

    def on_epoch_end(self):
        'Updates indexes after each epoch'
        self.indexes = np.arange(len(self.dataset_df))
        if self.shuffle == True:
            np.random.shuffle(self.indexes)

    def __len__(self) :
        return int(np.floor(len(self.dataset_df) / float(self.batch_size)))

    def __getitem__(self, index):
        # Generate indexes of the batch
        indexes = self.indexes[index*self.batch_size:(index+1)*self.batch_size]
        # Find list of IDs
        #images_filenames_temp = [self.image_filenames[k] for k in indexes]
        # Generate data
        X, y = self.__data_generation(indexes)

        return X, y

    def __data_generation(self, indexes):
        'Generates data containing batch_size samples'
        # Initialization
        x = np.empty((self.batch_size, *self.target_size, 3))
        y = np.empty((self.batch_size, len(classes)))

        # Generate data
        for i, index in enumerate(indexes):
            filename = self.dataset_df['source_file'][index]
            temp_img = cv2.imread(filename)

            crop_columns = ['x1_crop', 'y1_crop', 'x2_crop', 'y2_crop']
            x1,y1,x2,y2 = self.dataset_df[crop_columns].iloc[index,:]

            if x1 != -1:
                h = y1-y1
                w = x2-x1
                if h < w:
                    cy = (y1+y2)//2
                    y1 = cy - w//2
                    y2 = cy + w//2
                    y1 = max(y1, 0)
                    y2 = min(y2, temp_img.shape[0])
                elif w < h:
                    cx = (x1+x2)//2
                    x1 = cx - h//2
                    x2 = cx + h//2
                    x1 = max(x1, 0)
                    x2 = min(x2, temp_img.shape[1])

                temp_img = temp_img[y1:y2, x1:x2]
                m_w = max(y2-y1, x2-x1)
                top = (m_w - (y2-y1)) // 2
                bottom = m_w - (y2-y1) - top
                left = (m_w - (x2-x1)) // 2
                right = m_w - (x2-x1) - left
                if not (top == 0 and bottom == 0 and left == 0 and right == 0):
                    temp_img = cv2.copyMakeBorder(temp_img, top, bottom, left, right, cv2.BORDER_REPLICATE)

            if temp_img.shape != self.target_size:
                temp_img = cv2.resize(temp_img, self.target_size)

            temp_img = np.array(temp_img).astype('float32')

            if self.need_normalize:
                temp_img = temp_img/255.0

            # Store sample
            x[i,] = temp_img
            # Store class
            y[i,] = utils.to_categorical(class_to_int[self.dataset_df['class'][index]], num_classes=len(classes))
        return x, y
```

```python
train_params = {
    "target_size": (100,100),
    "batch_size": 32,
    "shuffle": True,
    "need_normalize": True,
}

test_params = dict(train_params)
test_params['batch_size'] = 1
test_params['shuffle'] = False

train_gen = My_File_Generator(datasets_united_train, **train_params)
dataset1_val_gen = My_File_Generator(dataset1_val, **test_params)
dataset1_test_gen = My_File_Generator(dataset1_test, **test_params)
dataset2_val_gen = My_File_Generator(dataset2_val, **test_params)
dataset2_test_gen = My_File_Generator(dataset2_test, **test_params)
```

```python
# Defining the CNN

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Conv2D(32, (5, 5), padding='same', activation='relu', input_shape=(100, 100, 3)))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
model.add(tf.keras.layers.Conv2D(64, (5, 5), padding='same', activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
model.add(tf.keras.layers.Conv2D(128, (5, 5), padding='same', activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
model.add(tf.keras.layers.Conv2D(256, (5, 5), padding='same', activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(256, activation='relu'))
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(len(classes), activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
save_best_cb = tf.keras.callbacks.ModelCheckpoint('models/initial-end-to-end', save_best_only=True) # Saves the best model so far
early_stopping_cb = tf.keras.callbacks.EarlyStopping(patience=5) # Interupts training when there is no progress
```

```python
history = model.fit(train_gen,
                    validation_data=dataset2_val_gen,
                    use_multiprocessing=False,
                    #workers=16,
                    epochs=20, callbacks=[save_best_cb, early_stopping_cb])
```

```python
def nn_predict(model, gen):
    probits = model.predict(gen)
    pred_int = probits.argmax(axis=1)
    return pred_int
```

```python
def get_y_from_df(df):
    #x = df.iloc[:, df.columns.get_loc('x0'):].to_numpy()
    y = df['class'].map(class_to_int).to_numpy()

    return y
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
```

```python
def nn_evaulate_on_dataset(model, gen, df, keyword):
    metrics = {'keyword': keyword}

    y_true = get_y_from_df(df)
    y_pred = nn_predict(model, gen)

    metrics['accuracy'] = accuracy_score(y_true, y_pred, normalize=True)
    metrics['f1'] = f1_score(y_true, y_pred, average="macro")
    metrics['precision'] = precision_score(y_true, y_pred, average="macro")
    metrics['recall'] = recall_score(y_true, y_pred, average="macro")

    return metrics
```

```python
def unite_results_by_keywords(*all_results):
    result = {}
    for candidate in all_results:
        key = candidate['keyword']
        for k, v in candidate.items():
            if k == 'keyword':
                continue
            result[key+'_'+k] = v
    return result
```

```python
all_results = unite_results_by_keywords(results_dataset1_val, results_dataset2_val, results_dataset1_test, results_dataset2_test)
```
